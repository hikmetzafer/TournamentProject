```python
import random

def footmatch(home, away):
        
    home_total, away_total = 0,0

    for matches in range(2):

        list_random = []
        goal_try = random.randint(0,10)
        home_score,away_score = 0,0

        while goal_try < random.randint(0,100):
            list_random.append(random.randint(0,1000))
            goal_try += 3

        for goal in list_random:
            if goal % 3 == 1:
                home_score += 1
            elif goal % 3 == 2:
                away_score += 1

        if matches < 1:
            print(f"\n{matches+1}. match's score is {home} {home_score} : {away} {away_score}")
        else :
            print(f"\n{matches+1}. match's score is {away} {away_score} : {home} {home_score}")

        home_total += home_score
        away_total += away_score

    if home_total == away_total:
        penalties = random.choice([True, False])
        if penalties:
            result = f'Draw. {home} won with penalty shots!'
            winner = home
        else:
            result = f'Draw. {away} won with penalty shots!'
            winner = away
    elif home_total > away_total:
        result = f'{home} won!'
        winner = home
    else:
        result = f'{away} won!'
        winner = away
    print('\n'+result, end='\n\n')
    return winner

print("You should enter 4 teams for the Tournament\n")

i = 1
list_teams = []    

while i < 5:
    team = input(f"{i}. team of 4").upper()
    list_teams.append(team)
    i += 1   

print(f"\nTournament teams are : {list_teams}")

team_1 = list_teams.pop(random.randint(0,3)) 
team_2 = list_teams.pop(random.randint(0,2))
team_3 = list_teams.pop(random.randint(0,1))
team_4 = list_teams[0]

print(f"\nDRAWS\n\n{team_1} - {team_2} \n{team_3} - {team_4}\n")



print(f"------------------------------\nSemi-final {team_1} - {team_2}")
final_1 = footmatch(team_1, team_2)
print(f"------------------------------\nSemi-final {team_3} - {team_4}")
final_2 = footmatch(team_3, team_4)
print(f"------------------------------\nFINAL {final_1} - {final_2}")
champion = footmatch(final_1, final_2)

print(f"The CHAMPION of the tournament is >>>> {champion}\nCongratulations {champion} !!!")
```
